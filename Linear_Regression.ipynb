```python
base_casas.columns = base_casas.columns.str.replace(' ', '_')
base_casas.head()
```

```python
base_casas.drop(["Address"], axis=1, inplace=True) # axis = 1 elimina uma coluna, = 0 uma linha. inplace = true, significa que faremos isso na base original
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = base_casas[['Avg._Area_Income', 'Avg._Area_House_Age', 'Avg._Area_Number_of_Rooms','Avg._Area_Number_of_Bedrooms', 'Area_Population']]
Y = base_casas[["Price"]]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, train_size = 0.7, test_size = 0.3, random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train, Y_train)
```

agora tentamos predizer esse valor na base de teste:


```python
y_pred = lm.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

o r2 score é um parametro que quanto maior, maior a possibilidade de acertar a regressão linear

```python
r = r2_score(Y_test, y_pred)
```

```python
print("R quadrado: ", r)
```

O modelo possui alta linearidade

Significa que quando as variáveis explicativas mudarem, temos 91% de chances da variável y mudar.
