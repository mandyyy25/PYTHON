#Grafics Using *SEABORN*
I made some plot of statistical data using this library.

```python
import pandas as pd
```

```python
data_houses = pd.read_excel("/content/houses_rent.xlsx")
```

```python
data_houses.head(10)
```

```python
data_houses = data_houses.rename(columns={"parking spaces":"parking_spaces"})
```

```python
data_houses.head(10)
```

```python
data_houses.dtypes
```

```python
data_houses = data_houses.astype({"rent amount":"float"})
```

```python
data_houses.info()
```

```python
data_houses = data_houses[(data_houses["total"]<=4000)]
data_houses.head()
```

```python
data_houses.describe()
```

```python
data_houses = data_houses[(data_houses["total"]<=4000) & (data_houses["animal"]=="accept")]
data_houses.head()
```

```python
data_houses = data_houses[(data_houses["total"]<=4000) & (data_houses["animal"]=="accept") & ((data_houses["rooms"]==2)| (data_houses["rooms"]==3))]
data_houses.head()
```

```python
data_houses.describe()
```

```python
data_houses_4000 = data_houses.query("total <= 4000")
data_houses_4000.head()
```

```python
agg_city_price1 = data_houses.groupby(["city"])["total"].agg("mean").reset_index()
agg_city_price1
```

```python
agg_city_price = data_houses.groupby(["bathroom", "rooms"])["total"].agg("mean").reset_index()
agg_city_price
```

```python
import seaborn as sns
```

```python
ax = sns.barplot(data = agg_city_price, x = "rooms", y = "total", color = "pink")
ax.set(xlabel = "rooms" , ylabel = "total")
```

```python
ax.set(xlabel = "rooms" , ylabel = "total")
```

```python
ax = sns.barplot(data = agg_city_price, x = "bathroom", y = "total", color = "green")
ax.set(xlabel = "bathrooms" , ylabel = "total")
```
